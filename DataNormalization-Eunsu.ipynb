```python
import os
import zipfile
from PIL import Image
import numpy as np

BASE_DIR = '/content/drive/MyDrive/25Spring/CIS630_Advanced_Artificial_Intelligence_Systems/Project'

# Define paths
zip_file_path = BASE_DIR +'/asl.zip'       # Path to your zip file
extract_dir = BASE_DIR +'/asl_extracted'   # Directory to extract images
output_dir = BASE_DIR +'/asl_data'         # Directory to save processed images

# Create directories if they don't exist
os.makedirs(extract_dir, exist_ok=True)
os.makedirs(output_dir, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# Process images in all subdirectories (0-9 and a-z)
for root, _, files in os.walk(extract_dir):  # Recursively iterate through all folders
    for filename in files:
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Check for image file types
            # Full path to the image
            img_path = os.path.join(root, filename)
            img = Image.open(img_path)

            # Convert to grayscale
            bw_img = img.convert('L')

            # Normalize the pixel values
            bw_array = np.array(bw_img) / 255.0  # Normalize to [0, 1]

            # Convert to black and white using threshold
            #bw_image = bw_img.point(lambda x: 0 if x < 128 else 255)

            # Maintain subfolder structure
            relative_path = os.path.relpath(root, extract_dir)  # Extract subfolder name (e.g., "a", "b", "0", etc.)
            save_folder = os.path.join(output_dir, relative_path)
            os.makedirs(save_folder, exist_ok=True)  # Create subfolder if it doesn't exist

            # Save the processed image in the corresponding subdirectory
            save_path = os.path.join(save_folder, filename)
            bw_img.save(save_path)

print("Unzipping and Normalization complete!")
```
